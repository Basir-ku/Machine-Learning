```python
# get the dummy variable 
dummy_sex = pd.get_dummies(df3.sex,drop_first=True, prefix="sex")
dummy_smoker = pd.get_dummies(df3.smoker,drop_first=True, prefix="smoker")
dummy_region = pd.get_dummies(df3.region,drop_first=True, prefix="region")

# now we need to replace sex, smoker, region variable with the dummy variable
# we can do it by droping the columns and add the dummy columns

# drop the columns 
df3 = df3.drop(["sex","smoker","region"], axis=1)
# Add df3 and dummy variable using concat functions passing list of variable/columns to concat
df3 = pd.concat([df3, dummy_sex, dummy_smoker, dummy_region],axis=1)

df3.head()


#dummy_region.head(2)
# dummy_sex.head()
# dummy_smoker.head(2)
```

```python
from sklearn.preprocessing import OrdinalEncoder

# get the uniques values for catagories
sex = df5.sex.unique()
smoker = df5.smoker.unique()
region = df5.region.unique()
#print(sex)
print(sex)
# crate objects
ordinal_sex = OrdinalEncoder(categories=[sex])
ordinal_smoker = OrdinalEncoder(categories=[smoker])
ordinal_region = OrdinalEncoder(categories=[region])
#print(ordinal_sex)

# fit and transform
encoded_sex = ordinal_sex.fit_transform(df5[["sex"]])
encoded_smoker = ordinal_smoker.fit_transform(df5[["smoker"]])
encoded_region = ordinal_region.fit_transform(df5[["region"]])
#print(encoded_sex)
```

```python
# conver ndarry to dataframe
df5["sex"] = pd.DataFrame(encoded_sex, columns=["sex"])
df5["smoker"] = pd.DataFrame(encoded_smoker, columns=["smoker"])
df5["region"] = pd.DataFrame(encoded_region, columns=["region"])
df5.head(2)
```
